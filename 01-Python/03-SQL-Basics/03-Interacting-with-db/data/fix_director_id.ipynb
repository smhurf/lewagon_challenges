```python

import sqlite3

conn = sqlite3.connect('movies.sqlite')
db = conn.cursor()
```

```python
# get all movies
movies = db.execute("SELECT id, director_id FROM movies")
movies = movies.fetchall()
movies
```

```python
for movie in movies:
    director_id = db.execute("SELECT id FROM directors WHERE imdb_director_id=?", (movie[1],))
    director_id = director_id.fetchone()[0]
    db.execute("UPDATE movies SET director_id=? WHERE id=?", (director_id,movie[0]))
```

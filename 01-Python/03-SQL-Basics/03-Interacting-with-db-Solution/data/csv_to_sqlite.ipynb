# CSV to SQLite

```python
import pandas as pd
import sqlite3
```

```python
conn = sqlite3.connect("movies.sqlite")
```

```python
conn.execute("CREATE TABLE if not exists movies (\
title TEXT,\
rating TEXT,\
vote_count INTEGER,\
start_year INTEGER,\
minutes INTEGER,\
genres TEXT,\
director_id TEXT,\
imdb_id TEXT,\
id INTEGER PRIMARY KEY AUTOINCREMENT\
)")

conn.execute("CREATE TABLE if not exists directors (\
name TEXT,\
birth_year INTEGER,\
death_year INTEGER,\
imdb_director_id TEXT,\
id INTEGER PRIMARY KEY AUTOINCREMENT,\
UNIQUE(imdb_director_id)\
)")
```

```python
cursor = conn.cursor()
cursor.execute("SELECT sql FROM sqlite_master")
cursor.fetchall()
```

```python
df = pd.read_csv("movies.csv")
```

```python
c = conn.cursor()

for index, row in df.iterrows():
    values = f'("{row["primaryTitle"]}", {row["averageRating"]}, {row["numVotes"]}, {row["startYear"]}, {row["runtimeMinutes"]}, "{row["genres"]}", "{row["nconst"]}", "{row["tconst"]}")'
    query = f'INSERT OR REPLACE INTO movies (title, rating, vote_count, start_year, minutes, genres, director_id, imdb_id) values {values}'
    c.execute(query)
    conn.commit()
    print(query)
c.close()
```

```python
c = conn.cursor()

for index, row in df.iterrows():
    values = f'("{row["primaryName"]}", {row["birthYear"]}, {row["deathYear"]}, "{row["nconst"]}")'
    query = f'INSERT OR REPLACE INTO directors (name, birth_year, death_year, imdb_director_id) values {values}'
    c.execute(query)
    conn.commit()
    print(query)
c.close()
```

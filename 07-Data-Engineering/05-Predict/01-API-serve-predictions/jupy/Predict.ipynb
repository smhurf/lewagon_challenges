# Get test data

```python
import pandas as pd
url = "gs://wagon-data/data/bq_complete.csv"
df = pd.read_csv(url, nrows=1000)
```

```python
df.shape
```

```python
df = df.dropna(how='any',  axis='rows')
instances = df.drop("fare_amount", axis=1).to_dict(orient="records")
```
